```python
import pandas as pd
import hvplot.pandas  # Import hvPlot to extend Pandas DataFrames for visualization
import panel as pn

pn.extension()

# Sample data for testing
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [23, 17, 35, 29, 12]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a bar chart function
def create_bar_chart(dataframe):
    return dataframe.hvplot.bar(x='Category', y='Values', title="Simple Bar Chart", width=800, height=400)

# Create the bar chart using the function
bar_chart = create_bar_chart(df)


# Create the Panel FastListTemplate including the bar chart
dashboard = pn.template.FastListTemplate(
    title="Test Dashboard with Bar Chart",
    sidebar=["# Sidebar", "Some information here..."],
    main=["# Simple Bar Chart Example", bar_chart]
).servable()
```
